## **ECUACIONES DE MAXWELL**

- *Las ecuaciones de Maxwell son  cuatro ecuaciones que describen el comportamiento de los campos electromagneticos en el espacio y el tiempo. Estas ecuaciones fueron desarrolladas por el físico escocés James Clerk Maxwell en la década de 1860 y son consideradas como una de las piedras angulares de la física moderna.*

*las ecucaiones de maxwell son:*

1. *ley de Gauss para el campo electrico:*

$\hspace{30mm}\vec{\nabla} \cdot   \vec{E} = \frac{\rho }{\varepsilon_0}$

2. *Ley de Gauus para el campo magnetico:*

$\hspace{30mm} \vec{\nabla} \cdot   \vec{B} = 0$

3. *Ley de Faraday para la induccion electromagnetico :*

$\hspace{30mm \vec{\nabla} \cdot   \vec{E} = - \frac{\partial \vec{B}}{\partial t}} $

4. *Ley de Ampere- Maxwell para la corriente electrica y campo magnetico:*

$\hspace{30mm \vec{\nabla} \cdot   \vec{E} = \mu_0 \left ( \vec{J} +  \varepsilon _0 \frac{\partial \vec{E}}{\partial t}   \right )}$

### ***1. LEY DE GAUSS PARA EL CAMPO ELECTRICO:***

*Esta escuacion establece que la divergencia del campo electrico en un punto es proporcional  a la densidad de carga electrica en ese punto. En otras palabras, la carga electrica crea un  campo electrico que se propaga en todas las direcciones y su magnitud disminuye con  la distancia. De manera que el campo electrico diverge en cargas porsitivas y converge en las negativas*

$$\vec{\nabla} \cdot   \vec{E} = \frac{\rho }{\varepsilon_0}$$

- ### ***Forma Diferencial:***

la primera Ecuacion esta  dada en forma diferencial como:

$\vec{\nabla} \cdot   \vec{E} = \frac{\rho }{\varepsilon_0}$

-*$\nabla$ es el operador nabla, que se define como $\nabla = \frac{\partial}{\partial x}\hat{x} + \frac{\partial}{\partial y}\hat{y} + \frac{\partial}{\partial z}\hat{z}$, y que representa la derivada espacial.*

*La ecuacion establece que  entre la densidad de carga y el campo eléctrico en un  cualquiera en el espacio y nos dice que la divergencia del campo eléctrico en un punto es igual a la densidad de carga en ese mismo punto dividida por la constante dieléctrica del vacío.*

*la ecuaion la podemos de motrar en base de la ley de Coulomb, que establece que la fuerza eléctrica entre dos cargas es proporcional al producto de las cargas e inversamente proporcional al cuadrado de la distancia entre ellas. A partir de esta ley, podemos demostrar que la divergencia del campo eléctrico es proporcional a la densidad de carga en un punto dado del espacio. y nos da para decir que la divergencia del campo electrico en un punto es igual a la suma de las cargas en una esfera  que se centra en este punto, dividida por la constante dieléctrica del vacío y por el volumen de la esfera. -*Donde $\mathbf{F}$ es la fuerza eléctrica, $\epsilon_0$ es la constante dieléctrica del vacío, $q_1$ y $q_2$ son las cargas, $r$ es la distancia entre ellas, y $\hat{r}$ es el vector unitario que apunta desde la carga $q_1$ hacia la carga $q_2$.*

*Ahora sustituimos las expresiones en la  divergencia*
$$ \nabla \cdot E =\frac{\partial }{\partial x}\left ( \frac{1}{4\pi\epsilon_0}\frac{q}{r^2}\frac{x-x_0}{r} \right ) + \frac{\partial }{\partial y}\left ( \frac{1}{4\pi\epsilon_0}\frac{q}{r^2}\frac{y-y_0}{r} \right ) + \frac{\partial }{\partial z}\left ( \frac{1}{4\pi\epsilon_0}\frac{q}{r^2}\frac{z-z_0}{r} \right )$$

- *Para faciliar los caculos y orden de sitaxis, creamos una funcion para calcular el campo electrico, la cual le metemos "r" para que el calulo se haga en un punto, tambien le damos lo valores para la posicion y carga puntual "qPos" y "rPos".*

- *Creamos una cuadricula en  plano x,y*

- *ya con la Cuadricula Cuadricula creada, y con la funcion calculamos el campo electrico en cada punto y las almacenamos en  en las matrices  "E_x"  y "E_y".*

- *Finalmente graficamos el campo electrico , lo cual va representar las flechas en cada punto de la cuadricula que representan la direcciony magnitud del campo electrico en ese punto.*



```python
#@utor : Edward Fabian Goyeneche.
# Date : 24-03-2023
# Subjet: Electromagnetic Fields  

import numpy as np
import matplotlib.pyplot as plt

epsilon0 = 8.85418782e-12  
q = 1.60217662e-19  

def electricField(r, qPos, rPos):

    dist = np.linalg.norm(r - rPos)
    E = qPos / (4 * np.pi * epsilon0 * dist**2) * (r - rPos) / dist
    return E

x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)

# Definimos la posición y carga de una carga puntual positiva en el origen
qPos = 1 * q
rPos = np.array([0, 0, 0])


# Calculamos el campo eléctrico en cada punto.
E_x = np.zeros_like(X)
E_y = np.zeros_like(Y)
for i in range(len(x)):
    for j in range(len(y)):
        r = np.array([x[i], y[j], 0])
        E = electricField(r, qPos, rPos)
        E_x[i,j] = E[0]
        E_y[i,j] = E[1]

E_mag = np.sqrt(E_x**2 + E_y**2)
E_x_norm = E_x / E_mag
E_y_norm = E_y / E_mag


fig, ax = plt.subplots()
ax.quiver(X, Y, E_x_norm, E_y_norm, E_mag, cmap='cool', alpha=0.8)
ax.set_xlabel("X (m)")
ax.set_ylabel("Y (m)")
plt.show()
```

*La gráfica muestra el campo eléctrico producido por una carga puntual positiva en el origen en un plano "x,y". Las flechas indican la dirección y magnitud del campo eléctrico en cada punto de la cuadrícula. El color de las flechas refleja la magnitud del campo eléctrico, el color azul celestes indica valores más bajos y el color violeta valores más altos.*

*Podemos ver que las flechas apuntan hacia afuera desde la carga positiva en el origen, lo que indica que el campo eléctrico es radial y se propaga en todas las direcciones. La magnitud del campo eléctrico disminuye a medida que nos alejamos de la carga, como se puede observar por el cambio de color de las flechas a medida que nos alejamos del origen. También podemos observar que el campo eléctrico es más fuerte cerca de la carga y más débil a medida que nos alejamos de ella, lo que es consistente con la ley de Coulomb.*


## **Glosario de Terminos**:

- *Campo eléctrico $(E)$: es la fuerza eléctrica por unidad de carga que actúa sobre una partícula cargada en un punto dado del espacio. Su unidad es newton por coulomb $(N/C)$.*

- *Divergencia $(∇·E)$: es una medida de cómo el campo eléctrico se distribuye en el espacio. La divergencia de un campo eléctrico en un punto dado es la cantidad de flujo eléctrico que entra o sale de ese punto por unidad de volumen. Su unidad es coulomb por metro cúbico $(C/m^3)$*.

- *Carga eléctrica $(q)$: es una propiedad fundamental de las partículas subatómicas que lleva una carga eléctrica. La carga eléctrica se mide en coulombs $(C)$*.

- *Permitividad del vacío (ε0): es una constante física que describe la capacidad del espacio vacío para soportar un campo eléctrico. Su valor es de aproximadamente $8.85 x 10^-12 F/m$*.

- *Volumen $(V)$: es una magnitud que se refiere al espacio tridimensional ocupado por un objeto. Su unidad es metro cúbico $(m^3)$*.

- *Superficie $(S)$: es una magnitud que se refiere a la extensión plana de un objeto. Su unidad es metro cuadrado $(m^2)$*.

- *Flujo eléctrico $(ΦE$: es una medida del número de líneas de campo eléctrico que pasan a través de una superficie dada. Su unidad es voltios por metro $V/m)$ o weber $(Wb)$*.

### ***LEY DE GAUSS PARA EL CAMPO MAGNETICO:***
*La segunda ecuación de Maxwell es la Ley de Gauss para el campo magnético, que establece que el flujo magnético neto a través de cualquier superficie cerrada es siempre cero..*

$$ \vec{\nabla} \cdot   \vec{B} = 0$$
- ### ***Forma Diferencial:***
*Matematicamente se expresa como:*
$$\nabla  \cdot E = 0$$

*donde $\nabla \cdot$ es el operador de divergencia, $\mathbf{B}$ es el campo magnético y $0$ representa la ausencia de fuentes magneticas en la region.*
*Empezando con la demostracion y explicacion en forma diferencial.Ponemos la definicion del flujo magnetico a tracez de una superficie cerrada $S$.*

$$\Phi_B= \iint_{S} B \cdot dS$$

*donde $dS$ es un elemento diferencial de área en la superficie $S$.*

-*Ahora aplicamos el teorema de la divergencia de Gauss para convertir la integral de superficie en una integral de volumen:*

$$\Phi_B=\int \int \int_V \nabla \cdot B dV $$

*donde $V$ es el volumen encerrado por la superficie $S$.*

-*Se utiliza el principio de conservación de la carga para afirmar que la carga total encerrada en el volumen $V$ es constante. Esto implica que la carga neta dentro de $V$ es cero*

$$\int \int \int_V  \rho dV = 0$$

*donde $\rho$ es la densidad de carga.*

-*Aplicamos la ley de Ampere-Maxwell, de este mismo grupo de ecuacion y tomamos la definicion de esta, que establece que la corriente electrica es la fuente de campo magnetico, para relacionar la densidad de carga con la corriente:*

$$\rho = \frac {\nabla \cdot J}{\mu_ 0}$$

*donde $J$ es la densidad de corriente y $\mu_0$ es la permeabilidad magnética del vacío.*

-*Tomamos la ecucion de la densidad de carga en la ecuacion de conservacion de carga:*
$$\iiint_{V} \frac{\nabla \cdot J}{\mu_ 0} dV= 0$$

-*Aplicamos el teorema de la divergencia de Gauss para convertir la integral de volumen en integral de superficie.*
$$\iint_{S} \frac{J \cdot dS}{\mu_ 0} = 0$$

-*Relacionamos la densidad de corriente con el campo magnetico, eso utilizando Ampere_Maxwell*
$$\oint_C B \cdot dI = \mu_0 \iint_{S} J \cdot dS$$

*donde $C$ es cualquier curva cerrada que encierra la superficie $S$.*

-*ahora sustitucimos la integral de la densidad de corriente en la ecuciones del anterior paso.*
$$\oint_C B \cdot \mathrm{d}\mathbf{l} = \mu_0 \iint_S \nabla \times \mathbf{B} \cdot \mathrm{d}\mathbf{S}$$

-*Utilizamos la propiedad de que el flujo magnetico a traves de  cualquier superficie cerrada es igual a $"0"$ para afirmar que la integral de superficie de la ecuacion anterio llega a dar $"0"$*

$$\iint_{S} \nabla \times B \cdot dS = 0$$

-*Recurrimos a usar el teorema de Stokes para convertir la integral de superficie en un integral de controno.*
$$\oint_C B \cdot dI =0 $$

-*Finalmente la integral e contorno de la ecuacion es siempre $"0"$ para cualquier curva cerrada $C$, lo que implica que el magnetico es soleidal.*

$$\nabla \cdot B = 0$$




- ### ***Forma  Integral:***
*En forma integral usamos de nuevo la ley de Ampere-Maxwell, peroya que esta va estar aplicada a una superficie cerrada en lugar de una curva cerrada.*

-*Esta ecuacion en esta forma relaciona la circulacion de campo eletrico a lo largo de una curva cerrad con la variacion temporal del flujo magnetico a traves de la superficie que esta curva encierra.*

$$\oint _C E \cdot dI = -\frac{d}{dt} \iint_{S} B \cdot dS$$

*donde $C$ es una curva cerrada que encierra una superficie $S$, $\mathbf{E}$ es el campo eléctrico, $\mathbf{B}$ es el campo magnético y $\mathrm{d}\mathbf{l}$ y $\mathrm{d}\mathbf{S}$ son elementos de línea y superficie*

-*Ahora aplicamos la ley de faraday , que relaciona la circulacion del campo electrico a lo largo de una curva cerrada con la variacion temporal del flujo electrico a traves de la superficie que la curva llega a encerrar.*

$$\oint _C E \cdot dI = -\frac{d}{dt} \iint_{S} \frac{\partial B}{\partial t} \cdot dS$$

-*Igualamos la ecuaciones obtenidas en los pasos anteriores y simplicamos lo  mas que se pueda los terminos.*

$$\iint_{S}(\nabla \times B) \cdot dS = -\mu_0 \iint_{S} \frac{\partial E}{\partial t}\cdot  dS$$

*Aqui utiliamos un rotor en terminos de corrdenadas cartersianas.*

-*Aplicamos la ley de Gauss para el campo electrico, que relaciona el flujo electrico a traves de una superficie cerra con la carga encerrada por esta superficie.*

$$\iint_{S}(\nabla \cdot E \cdot dS =\frac{Q_enc}{\varepsilon _0}$$

*donde $Q_{enc}$ es la carga encerrada por la superficie $S$ y $\epsilon_0$ es la permitividad eléctrica del vacío.*

-*Consecuente  usamos la ley de continuidad de la carga, que establece que la carga no puede se crea ni se destruye, por lo tanto caulquier cambio en la carga en una region del espacio debe ser debido  a la corriente que fluye dentro o fuera de esta region.*

$$\nabla \cdot J + \frac{\partial \rho}{\partial t} = 0$$

*donde $J$ es la densidad de corriente eléctrica y $\rho$ es la densidad de carga.*

-*Aplicamos el teorema de la dicergencia de la ley de Gauss a la ecuacion obtenida en el paso 5.*

$$\iint_{S}\nabla \cdot J \cdot dS + \frac{d}{dt}\iiint_{V} \cdot dv = 0$$
*donde $V$ es el volumen encerrado por la superficie $S$*

-*Reemplazamos en la ecuacion obtenidas  obtenidas anteriormente*
$$\iint_{S}\nabla \cdot J \cdot dS + \frac{d}{dt}\iint_{V} \frac{\nabla \cdot E}{\varepsilon_0}\cdot dS = 0$$

-*Utilizamos la ley de Ampere-Maxwell en forma integral para el termino $\nabla \cdot \mathbf{J}$ en términos del campo magnético.*

$$\iint_{S}(\nabla \times B)\cdot dS = \mu _0 \iint_{S} J \cdot dS + \mu_0 \varepsilon_0 \frac{d}{dt} \iint_{S}E \cdot dS$$

-*Reemplazamos la ecuccion obtenia en la ecuacion anterior.*
$$\mu _0 \iint_{S} J \cdot dS + \mu_0 \varepsilon_0 \frac{d}{dt} \iint_{S}\frac{\partial E}{\partial t}\cdot dS +\frac{d}{dt}\iint_{S}\frac{\nabla \cdot E}{\varepsilon_0}\cdot dS = 0$$

-*Se utiliza la Ley de Gauss para el campo eléctrico para expresar el término $\nabla \cdot \mathbf{E}$ en términos de la densidad de carga*
$$\iint_{S} \frac{\nabla \cdot E}{\varepsilon_0}\cdot dS = \frac{Q_{enc}}{\varepsilon_0} $$

-*ahora reemplazamos la ecuacion obtenida en la obtenida anteriormente.*
$$\mu _0 \iint_{S} J \cdot dS + \mu_0 \varepsilon_0 \iint_{S} \frac{\partial E}{\partial t} \cdot dS+ \frac{d}{dt}\frac{Q_{enc}}{\varepsilon _0}= 0$$

-*Utilizamos la definicion del campo magnetico en terminos de la ley de Ampere-Maxwell en forma integral:*
$$\iint_{S}(\nabla \times  B) \cdot dS = \mu_0 \iint_{s} (\nabla \times B)\cdot dS + \mu_0 \varepsilon_0 \frac{d}{dt}\iint_{S}E \cdot dS$$

-*Finalmente reorganizamos la ecuacion  para que nos de el resultado final:*

### ***LEY DE  FARADAY PARA LA INDUCCION ELECTROMAGNETICA :***

*Esta ecuación establece que cualquier variación en el campo magnético en el tiempo produce un campo eléctrico en el espacio circundante. En otras palabras, la variación del campo magnético produce un flujo de cargas eléctricas que generan un campo eléctrico en la región.*

$$ \nabla \times \vec{E} = -\frac{\partial \vec{B} }{\partial t}  $$

- ### ***Forma Diferencial:***

-*En forma diferencial La tercera ecuación de Maxwell, también conocida como la Ley de Faraday, establece que un cambio en el flujo magnético a través de una superficie cerrada genera una fuerza electromotriz (fem) en cualquier circuito que esté atravesando dicha superficie. Matemáticamente,*

$$ \nabla \times \vec{E} = -\frac{\partial \vec{B} }{\partial t} $$

*Donde $\nabla \times \vec{E}$ es el rotacional del campo eléctrico $\vec{E}$, $\vec{B}$ es el campo magnético y $\frac{\partial \vec{B}}{\partial t}$ es la tasa de cambio temporal del campo magnético.*

-*Se puede comenzar con la ley de Faraday para una espira de alambre en un campo magnético uniforme.*

$$\varepsilon = - \frac{d \phi}{dt}$$

*Donde $\mathcal{E}$ es la fuerza electromotriz inducida, $\Phi$ es el flujo magnético a través de la espira y $\frac{d\Phi}{dt}$ es la tasa de cambio temporal del flujo magnético.*

- *Ahora  consideramos na superficie cerrada arbitraria $S$ en el espacio que está atravesada por un campo magnético variable en el tiempo. Podemos dividir esta superficie en un conjunto de pequeñas espiras de alambre, cada una con su propia fem $\mathcal{E}_i$ y flujo magnético $\Phi_i$. La suma de todas estas fems nos dará la fem total en el circuito, y la suma de todos los flujos magnéticos nos dará el flujo magnético total a través de la superficie cerrada.*

$$\oint_{C} \vec{E} \cdot d\vec{I} = -\frac{d}{dt}\iint_{S} \vec{B} \cdot d\vec{a}$$

*Donde $\oint_C$ es una integral de línea alrededor del borde de la superficie $S$ y $\iint_S$ es una integral de superficie sobre la superficie $S$. Aplicando el teorema de Stokes al lado izquierdo de la ecuación, podemos convertir la integral de línea en una integral de superficie del rotacional de $\vec{E}$ sobre la superficie $S$*

$$\iint_{S}(\nabla \times \vec{E})\cdot  d\vec{a} =-\frac{d}{dt}\iint_{S} \vec{B}\cdot d\vec{a}$$

-*Usando la regla de la cadena derivamos la integral de superficie del flujo magnetico.*


$$\iint_{S}(\nabla \times \vec{E})\cdot  d\vec{a} =-\frac{d}{dt}\iint_{S} \vec{B}\cdot d\vec{a}$$

-*finalmente  igualamos  las dos ecuaciones para que quede en forma diferencial.*

$$ \nabla \times \vec{E} = -\frac{\partial \vec{B} }{\partial t} $$



- ### ***Forma  Integral:***
-*En forma integral establece que la fem inducida en una superficie cerrada es igual a la tasa de cambio temporal del flujo magnético a través de esa superficie. En términos matemáticos.*

$$\int_{\partial S} E \cdot d\iota  = -\frac{d}{dt}\iint_{S} B \cdot da$$

*Donde $\partial S$ es el borde de la superficie cerrada $S$, $\mathbf{E}$ es el campo eléctrico, $\mathbf{B}$ es el campo magnético, $\mathrm{d}\boldsymbol{\ell}$ es el elemento de línea infinitesimal a lo largo del borde de $S$ y $\mathrm{d}\mathbf{a}$ es el elemento de área infinitesimal sobre $S$.*

-*Consideremos una superficie cerrada arbitraria $S$ en el espacio que está atravesada por un campo magnético variable en el tiempo. El flujo magnético a través de esta superficie cerrada se puede calcular.*

$$\Phi_B=\iint_{S} B \cdot da $$

-*luego suponemos qque el flujo magnético a través de la superficie cerrada $S$ cambia en el tiempo. Entonces, por la Ley de Faraday, una fem $\mathcal{E}$ se induce en cualquier circuito que atraviesa la superficie $S$. La fem total $\mathcal{E}_{tot}$ inducida en el circuito se puede encontrar integrando la fuerza electromotriz en todas las pequeñas espiras de alambre a través de la superficie $S$.*

$$\varepsilon _{tot} = \oint_{\partial S}E \cdot d\imath $$

*De lo anterior podemos escribir la ley de Faraday en terminos de la fem inducida en el circuito como*

$$\varepsilon _{tot} = - \frac{d \Phi_B}{dt}$$

*Finalmente Reemplazando $\Phi_B$ con su expresión integral y aplicando el teorema de Stokes al lado izquierdo de la ecuación, obtenemosla  3ra ecuacion de maxwell en su forma integral.*

$$\oint_{\partial S}E  \cdot d\imath  = -\frac{d}{dt}\iint_{S} B \cdot da$$

- ### ***Fisica:***

*Esta ecuacion  fisicamente establece que un cambio en el campo magnetico en el tiempo que genera un campo magnetico en el tiempo genera un campo electrico circulante. para entender esto como  ejemplo es un imán y una bobina de alambre cercanos. Cuando el imán se mueve hacia o lejos de la bobina, el campo magnético que atraviesa la bobina cambia en el tiempo. Este cambio en el campo magnético induce una fem (fuerza electromotriz) en la bobina, lo que significa que se genera una diferencia de potencial eléctrico en la bobina que puede ser aprovechada para hacer fluir una corriente eléctrica.*

*Este efecto se puede entender en términos de la ley de Lenz, que establece que la dirección de la fem inducida en una bobina es tal que se opone al cambio en el campo magnético que la genera. Esto significa que cuando el imán se mueve hacia la bobina, el campo magnético que atraviesa la bobina aumenta, lo que induce una fem en la bobina que se opone al aumento del campo magnético. De manera similar, cuando el imán se aleja de la bobina, el campo magnético que atraviesa la bobina disminuye, lo que induce una fem en la bobina que se opone a la disminución del campo magnético.*

*En conclusion, a tercera ecuación de Maxwell o la Ley de Faraday para la inducción electromagnética describe cómo un cambio en el campo magnético en el tiempo genera un campo eléctrico circulante. Este efecto es fundamental para entender cómo se puede generar una corriente eléctrica a partir de un campo magnético variable en el tiempo, lo que es esencial para el funcionamiento de dispositivos electromagnéticos como los generadores eléctricos.*

### ***LEY  DE AMPERE-MAXWELL PARA LA CORRIENTE ELECTRICA Y EL CAMPO MAGNETICO:***
*Esta ecuación establece que cualquier variación en el campo eléctrico produce un campo magnético en el espacio circundante. Además, este campo magnético está relacionado con la corriente eléctrica presente en la región. En otras palabras, la variación del campo eléctrico produce un flujo de cargas eléctricas que generan un campo magnético en la región.*

$$\hspace{30mm \vec{\nabla} \cdot   \vec{E} = \mu_0 \left ( \vec{J} +  \varepsilon _0 \frac{\partial \vec{E}}{\partial t}   \right )}$$

- ### ***Forma Diferencial:***
*Esta dada :*
$\nabla \times B = \mu_ 0 (J + \varepsilon _0 \frac{\partial E}{\partial t})$

*donde $\nabla \times \mathbf{B}$ es el rotacional del campo magnético $\mathbf{B}$, $\mu_0$ es la permeabilidad magnética del vacío, $\mathbf{J}$ es la densidad de corriente eléctrica y $\epsilon_0$ es la permitividad eléctrica del vacío.*

-*Comensamos con la expresion de la ley de Ampére*
$$\nabla \times B = /mu_0 J$$

-*Despues tomamos el rotacional de la ecuacion de Maxwell-Faraday.*
$$\nabla \times (\nabla \times E) = -\frac{\partial }{\partial t}\nabla \times B$$

-*Utilisamos la identidad vectorial $\nabla \times \left(\nabla \times \mathbf{A}\right) = \nabla\left(\nabla \cdot \mathbf{A}\right) - \nabla^2 \mathbf{A}$, podemos simplificar la ecuacion.*

$$\nabla  (\nabla \cdot  E)- \nabla^2 E = -\frac{\partial }{\partial t}(\nabla \times B)$$

-*teniendo en cuento,$\nabla \cdot \mathbf{E} = \frac{\rho}{\epsilon_0}$, donde $\rho$ es la densidad de carga eléctrica, dado asi podemos reescribir la ecuacion anterior.*
$$\nabla \left ( \frac{\rho}{\varepsilon _0} \right )- \nabla^2 E = -\frac{\partial }{\partial t}(\nabla \times B$$ 

*luego tomamos el rotacional e ambos lados de la ecuación, podemos utilizar las propiedades de la operación rotacional para simplificar la expresión*
$$ \nabla \times \left ( \nabla \left ( \frac{\rho}{\varepsilon _0} \right ) \right )-(\nabla^2E) = -\frac{\partial }{\partial t}(\nabla \times B)$$

-*Como anteriormente, tenemos que como $\nabla \times \left(\nabla f\right) = \mathbf{0}$ para cualquier función escalar $f$, podemos eliminar el primer término de la ecuacion.*
$$ - \nabla\times (\nabla ^2E)= -\frac{\partial }{\partial t}(\nabla \times B)$$

-*Utilizamos la identidad Vectrorial $\nabla \times \left(\nabla \times \mathbf{A}\right) = \nabla\left(\nabla \cdot \mathbf{A}\right) - \nabla^2 \mathbf{A}$ nuevamente, podemos escribir el lado izquierdo de la ecuación*

$$-\nabla (\nabla \times E) + \nabla ^2E = \nabla \left ( -\frac{\rho}{\varepsilon _0} \right ) + \nabla^2 E $$

-*Reescribiendo la ecuacion completa , esta queda.*
$$\nabla \left ( - \frac{\rho}{\varepsilon _0} \right )+ \nabla ^2E = -\frac{\partial }{\partial t}(\nabla \times B)$$

-*Pero como $\nabla \cdot \left(\mathbf{J}+\epsilon_0\frac{\partial \mathbf{E}}{\partial t}\right) = \frac{\partial \rho}{\partial t} + \epsilon_0 \nabla \cdot \frac{\partial \mathbf{E}}{\partial t} + \nabla \cdot \mathbf{J}$, podemos reemplazar el término $\nabla \left(-\frac{\rho}{\epsilon_0}\right)$ por $\nabla \cdot \left(\mathbf{J}+\epsilon_0\frac{\partial \mathbf{E}}{\partial t}\right)$*

$$ \nabla \cdot \left ( J +\varepsilon _0  \frac{\partial }{}\right )+ \nabla ^2 E = -\frac{\partial }{\partial t}(\nabla \times B) $$

-*Y finalmente, utilizando la identidad vectorial $\nabla \cdot \left(\nabla \times \mathbf{A}\right) = 0$ para cualquier campo vectorial $\mathbf{A}$, podemos eliminar el término $\nabla \cdot \left(\mathbf{J}+\epsilon_0\frac{\partial \mathbf{E}}{\partial t}\right)$*

$$ \nabla \times B = \mu_0 \left ( J +\varepsilon _0 \frac{\partial E}{\partial t} \right )$$





- ### ***Forma  Integral:***
-*Esta ecuacion en esta forma Relaciona la circulación del campo magnético alrededor de una curva cerrada con la corriente eléctrica que atraviesa la superficie que encierra dicha curva. Esta ley se puede expresar en forma integral.*
$$\oint_{\partial S} B \cdot dI ) \mu_0 \iint_{S} \left ( J + \varepsilon_0 \frac{\partial E}{\partial t}  \right )\cdot dS$$

*donde $\partial S$ es la curva cerrada que encierra una superficie $S$, $dI$ es un elemento diferencial de longitud a lo largo de $\partial S$, $dS$ es un elemento diferencial de superficie de $S$, $B$ es el campo magnético, $J$ es la densidad de corriente eléctrica y $E$ es el campo eléctrico*

-*Según la Ley de Ampère, la circulación del campo magnético alrededor de $\Gamma$ es igual a la corriente eléctrica que atraviesa la superficie $S'$ que es limitada por $\Gamma$*
$$\oint_{T} B \cdot dI = \mu_0 I_{enc}$$

*donde $I_{\rm enc}$ es la corriente eléctrica que atraviesa $S'$. Podemos dividir la superficie $S'$ en dos partes, una que es común a $S$ y otra que no lo es. La parte común a $S$ se cancela en la expresión integral de la Ley de Ampère-Maxwell, y la parte que no es común a $S$ se anula cuando la longitud de $\Gamma$ tiende a cero, ya que la densidad de corriente es finita en cualquier punto de $C$. Por lo tanto, podemos escribir.*

$$\oint_{T} B \cdot dI = \mu_0 \iint_{S'}\left ( J + \varepsilon _0 \frac{\partial E}{\partial t} \right ) \cdot dS = \mu _0 \iint_{S}\left ( J + \varepsilon _0 \frac{\partial E}{\partial t} \right )\cdot dS$$

*donde hemos utilizado la relación $\mathbf{J}+\epsilon_0\frac{\partial \mathbf{E}}{\partial t} = \frac{\partial \mathbf{D}}{\partial t}$, y hemos aplicado el Teorema de Stokes para reescribir la integral n términos de la superficie $S$ que está limitada por la curva cerrada $\partial S$.*

-*Po lo tanto la Ley de Ampere-Maxwell en forma integral como:*
$$\oint_{\partial S} B \cdot dI = \lim_{n \rightarrow \infty }\oint_{T_n}B \cdot dI = \lim_{n \rightarrow \infty } \mu_0 \iint_{S_n} \left ( J \cdot \varepsilon _0 \frac{\partial E}{\partial t} \right )\cdot dS$$

- ### ***Fisica:***
*En concepto Fisico Estable ue el cambio temporal del campo eléctrico a través de una superficie cerrada es igual a la circulación del campo magnético alrededor de la frontera de la superficie cerrada, más la contribución de la corriente eléctrica que atraviesa la superficie.*

*Es muy importante ya que relaciona el campo eléctrico con el campo magnético y la corriente eléctrica, y permite entender la interacción entre estas tres entidades en un sistema electromagnético.*


*Para entender la demostración física de la Ley de Ampère-Maxwell, podemos imaginar un circuito formado por una fuente de corriente eléctrica y un conductor. La corriente eléctrica fluye por el conductor y crea un campo magnético a su alrededor, como se muestra en la siguiente imagen*



*La línea de corriente $C$ en el conductor es la frontera de la superficie $S$, y el campo magnético $\mathbf{B}$ se curva alrededor de la línea de corriente. Si trazamos una línea cerrada $\Gamma$ alrededor de la línea de corriente $C$, podemos aplicar la Ley de Ampère para obtener la circulación del campo magnético alrededor de la línea cerrada.*

*En particular, el cambio temporal del campo eléctrico a través de la superficie cerrada $S$ está relacionado con la contribución de la corriente eléctrica que atraviesa la superficie, ya que la variación temporal del campo eléctrico se relaciona con la corriente eléctrica a través de la Ley de Faraday.*

*Conluyendo podemos decir que esta ecuacion ( ley de Ampere-Maxwell) en terminos fisco es:*
- *La circulación del campo magnético alrededor de la línea cerrada $\Gamma$ es igual a la corriente eléctrica que atraviesa la superficie encerrada por la línea cerrada $\Gamma$.*

- *Además, la variación temporal del campo eléctrico a través de la superficie cerrada $S$ está relacionada con la contribución de la corriente eléctrica que atraviesa la superficie.*

- ### ***Ejemplo***
*Utilizando python , vamos a graficas el campo magnetico Generado por una corriente electrica que fluye a travez de un alambre recto y largod, utilizando la ecuacion que trabajamos*



```python
#@utor : Edward Fabian Goyeneche.
# Date : 26-03-2023
# Subjet: Electromagnetic Fields  

import numpy as np
import matplotlib.pyplot as plt

# Constantes
mu0 = 4 * np.pi * 10**-7
I = 1.0   # Corriente eléctrica en el cable
a = 0.01  # Radio del cable


def B_field(x, y, z):
    # Cálculo de la integral de la Ley de Ampère-Maxwell
    r = np.sqrt(x**2 + y**2 + z**2)
    integrand = mu0 * I / (2 * np.pi * r) * a**2 / (a**2 + z**2)
    B_theta = integrand * (x / r)
    B_phi = integrand * (y / r)
    B_z = mu0 * I / (2 * np.pi * r) * np.arctan2(z, np.sqrt(x**2 + y**2))
    
    # Devuelve el campo magnético en coordenadas cilíndricas
    return (B_theta, B_phi, B_z)

# Creación de la cuadrícula de puntos
x = np.linspace(-0.1, 0.1, 50)
y = np.linspace(-0.1, 0.1, 50)
X, Y = np.meshgrid(x, y)

# Cálculo del campo magnético en cada punto
B_x, B_y, B_z = np.zeros_like(X), np.zeros_like(Y), np.zeros_like(X)
for i in range(len(x)):
    for j in range(len(y)):
        B_theta, B_phi, B_z[i,j] = B_field(X[i,j], Y[i,j], 0)
        B_x[i,j] = B_theta
        B_y[i,j] = B_phi
        
# Graficar el campo magnético
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, np.zeros_like(X), B_x, B_y, B_z, length=0.02, normalize=True)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

*la gráfica generada ,muestra el campo magnético creado por un cable recto largo que lleva corriente eléctrica en el centro de un plano. La dirección y la magnitud del campo magnético se indican mediante flechas (quivers) en un sistema de coordenadas tridimensional.*

*En la gráfica, el eje X representa la coordenada X, el eje Y representa la coordenada Y y el eje Z representa la coordenada Z. Los puntos en la gráfica son la cuadrícula de puntos creadas y se utilizan para calcular el campo magnético en cada punto.*
*La dirección y magnitud del campo magnético en cada punto se indica mediante flechas de longitud proporcional al campo magnético en ese punto. Las flechas están normalizadas y tienen una longitud de 0.02 para facilitar la visualización.*

### ***ECUACION DE CONTINUIDAD (EN ELECTROMAGNETISMO):***

*a ecuación de continuidad en electromagnetismo establece que la carga eléctrica en un volumen cerrado no puede ser creada ni destruida, sino que se conserva. Esto se puede expresar matemáticamente como:*
$$\frac{\partial \rho}{\partial t} + \nabla \cdot \mathbf{J} = 0$$

*donde $\rho$ es la densidad de carga eléctrica y $\mathbf{J}$ es la densidad de corriente eléctrica.*

*La primera parte de la ecuación, $\frac{\partial \rho}{\partial t}$, describe la tasa de cambio de la densidad de carga eléctrica con respecto al tiempo. Si la carga está siendo creada o destruida en el volumen cerrado, entonces esta tasa de cambio no será cero.*

*La segunda parte de la ecuación, $\nabla \cdot \mathbf{J}$, describe la divergencia de la densidad de corriente eléctrica. Esto significa que si hay una cantidad neta de corriente entrando o saliendo del volumen cerrado, entonces la divergencia no será cero.*

*Si combinamos estas dos partes de la ecuación, podemos ver que si la carga eléctrica está siendo creada o destruida, entonces debe haber una corriente eléctrica neta entrando o saliendo del volumen cerrado para que la ecuación se satisfaga y la carga eléctrica se conserve.*

*Ahora, para demostrar matemáticamente la ecuación de continuidad, comenzamos con la ecuación de conservación general:*

$$\frac{\partial \rho}{\partial t} + \nabla \cdot \mathbf{J} = -\nabla \cdot \mathbf{F}$$

*donde $F$ es el flujo de carga eléctrica a través de la superficie del volumen cerrado.*

*Integrando esta ecuación sobre el volumen cerrado y aplicando el teorema de la divergencia, obtenemos:*

$$\int_V \frac{\partial \rho}{\partial t} dV + \oint_S \mathbf{J} \cdot d\mathbf{S} = -\oint_S \mathbf{F} \cdot d\mathbf{S}$$

*donde $V$ es el volumen cerrado y $S$ es su superficie.*

*asumimos que el volumen cerrado no tiene fuentes o sumideros de carga eléctrica, entonces la carga eléctrica en el volumen cerrado no cambia con el tiempo, es decir, $\frac{\partial \rho}{\partial t} = 0$. También podemos asumir que el flujo de carga eléctrica a través de la superficie del volumen cerrado es cero, es decir, $F = 0$.*

*Con estas suposiciones, la ecuación de conservación general se reduce a:*


$$\oint_S \mathbf{J} \cdot d\mathbf{S} = 0$$


*Esta seria la ecuacion  de comtinuidad en electromagnetismo*

### ***Fisica:***

*La ecuación de continuidad en electromagnetismo es una consecuencia de la ley de conservación de la carga eléctrica. Esta ley establece que la carga eléctrica no puede ser creada ni destruida, sino que se conserva. La ecuación de continuidad expresa esta ley en términos de la densidad de carga eléctrica y la densidad de corriente eléctrica.*

*La densidad de carga eléctrica, $\rho$, es una medida de la cantidad de carga eléctrica por unidad de volumen en un material o en el espacio libre. La densidad de corriente eléctrica, $\mathbf{J}$, es una medida de la cantidad de carga eléctrica que fluye a través de una unidad de área en un material conductor.*

*La ecuación de continuidad establece que la tasa de cambio de la densidad de carga eléctrica en un volumen cerrado debe ser igual y opuesta a la divergencia de la densidad de corriente eléctrica en ese volumen cerrado. En otras palabras, cualquier cambio en la densidad de carga eléctrica dentro del volumen cerrado debe ser igual y opuesto a cualquier flujo neto de carga eléctrica dentro o fuera del volumen cerrado.*

*Esta ecuación es muy importante en la teoría electromagnética ya que puede ser utilizada para describir muchos fenómenos eléctricos. Por ejemplo, se puede usar para describir la propagación de ondas electromagnéticas en el espacio libre o en un material conductor. También se puede usar para describir la distribución de cargas eléctricas en un capacitor o en un cable conductor.*


- ### ***Ejemplo:***

*Simular la evolucion temporal de la  densidad de carga electria en un cilindro cargado uniforme en su extremo inferior.*


```python

#@utor : Edward Fabian Goyeneche.
# Date : 28-03-2023
# Subjet: Electromagnetic Fields  

import numpy as np
import matplotlib.pyplot as plt

# Parámetros del cilindro
radius = 0.1   # Radio del cilindro en metros
height = 1.0    # Altura del cilindro en metros
charge_density = 1.0e-9    # Densidad de carga eléctrica en Coulombs por metro cúbico

# Parámetros de tiempo
t_start = 0.0    # Tiempo inicial en segundos
t_end = 1.0    # Tiempo final en segundos
dt = 0.01    # Paso de tiempo en segundos

# Discretización del espacio
dr = 0.01    # Paso radial en metros
dz = 0.01    # Paso axial en metros

# Creación de la malla
r = np.arange(0, radius + dr, dr)
z = np.arange(0, height + dz, dz)
R, Z = np.meshgrid(r, z)

# Inicialización de la densidad de carga eléctrica
rho = np.zeros_like(R)
rho[:, 0] = charge_density    # Cilindro cargado en el extremo inferior

# Creación del gráfico
fig, ax = plt.subplots()

# Animación de la densidad de carga eléctrica
for t in np.arange(t_start, t_end + dt, dt):
    # Cálculo de la tasa de cambio temporal de la densidad de carga eléctrica
    d_rho_dt = - np.gradient(np.gradient(rho, axis=0), axis=0) - np.gradient(np.gradient(rho, axis=1), axis=1)
    
    # Actualización de la densidad de carga eléctrica
    rho = rho + d_rho_dt * dt
    
    # Cálculo de la carga eléctrica total
    total_charge = np.sum(rho) * dr * dz * np.pi * radius ** 2
    
    # Creación del gráfico
    ax.clear()
    ax.set_title("Densidad de carga eléctrica en el cilindro\nTiempo: {:.2f} s, Carga total: {:.2e} C".format(t, total_charge))
    ax.set_xlabel("Radio (m)")
    ax.set_ylabel("Altura (m)")
    ax.contourf(R, Z, rho)
    ax.set_xlim(0, radius)
    ax.set_ylim(0, height)
    plt.pause(0.001)

plt.show()
```

*La Grafica muestra  la densidad de carga  electrica en un plano radial del cilindro, donde el eje  X  representa  el radio del cilindro y el eje Y representa la altura del cilindro. en este grafico l adensidad  de carga electrica se representa  en un mapa de colores,  donde los colores  mas claro en este caso mas cercano a la linea del eje Y indica una mayor densidad de carga electrica y los colores mas oscuros respresentan menor densidad electrica.*

### **1.4.)Ecuaciones  Constitutivas.**


*Las ecuaciones constitutivas son ecuaciones matemáticas que describen la relación entre las variables de entrada y de salida de un material o sistema en particular. Estas ecuaciones describen cómo el material o sistema responde a diferentes estímulos o cargas, como la temperatura, la presión, la tensión y la deformación.*

*Las ecuaciones constitutivas son fundamentales en la mecánica de materiales y la ingeniería de materiales, ya que permiten modelar el comportamiento de los materiales y predecir cómo se comportarán en diferentes situaciones. Por ejemplo, las ecuaciones constitutivas se utilizan en la ingeniería estructural para predecir la respuesta de una estructura a diferentes cargas y en la ciencia de los materiales para entender la relación entre la microestructura del material y sus propiedades macroscópicas.*

*Estas ecuaciones aplicadas al electromagnetimos, describen la relación entre los campos eléctricos y magnéticos y las fuentes que los generan. Estas ecuaciones son fundamentales para entender y analizar el comportamiento de los campos electromagnéticos en diferentes medios y condiciones. Las ecuaciones constitutivas básicas en el electromagnetismo son:*

***1. Ley de Coulomb***

*La ley de Coulomb describe la fuerza eléctrica entre dos cargas puntuales. Matemáticamente, se puede escribir como:*

$$ F  = K \cdot \frac{(q_1 - q_2)}{r^2}$$

*Donde $F$ es la fuerza eléctrica, $k$ es la constante de Coulomb, $q_1$ y $q_2$ son las cargas eléctricas de las partículas y $r$ es la distancia entre ellas.*


***2. Ley de Gauss:***

*La ley de Gauss establece que el flujo eléctrico a través de una superficie cerrada es proporcional a la carga eléctrica dentro de la superficie. Matemáticamente, se puede escribir como:*

$$\oint E \cdot dA = \frac{Q}{\varepsilon _ 0}$$

*Donde $E$ es el campo eléctrico, dA es un elemento de área en la superficie cerrada, $Q$ es la carga eléctrica dentro de la superficie y $\varepsilon _ 0$ es la permitividad eléctrica del vacío.*


***3. Ley de Faraday :***

*La ley de Faraday describe la inducción electromagnética, es decir, la generación de una corriente eléctrica en un circuito cerrado debido a una variación en el flujo magnético que atraviesa la superficie del circuito. Matemáticamente, se puede escribir como:*

$$\oint E \cdot  dl =  \frac{d \Phi B}{dt}$$

*Donde $E$ es el campo eléctrico, $dl$ es un elemento de longitud en el circuito cerrado, $\Phi B$ es el flujo magnético que atraviesa la superficie del circuito y  $\frac{d \Phi B}{dt}$ es la variación temporal del flujo magnético.*

***4. Ley de Ampere - Maxwell:***

*La ley de Ampere-Maxwell describe la relación entre el campo magnético y las corrientes eléctricas y los campos eléctricos que varían en el tiempo. Matemáticamente, se puede escribir como:*

$$ \oint B \cdot dl = \mu_0 \cdot \left ( I + \varepsilon _0 \cdot  \frac{d \Phi E}{dt}  \right ) $$


*Donde B es el campo magnético, dl es un elemento de longitud en la línea cerrada, $\mu _0$ es la permeabilidad magnética del vacío, $I$ es la corriente eléctrica, $\varepsilon_ 0$ es la permitividad eléctrica del vacío y $\frac{d \Phi E}{dt}$ es la variación temporal del flujo eléctrico a través de la superficie limitada por la línea cerrada.*
